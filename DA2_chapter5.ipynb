## **LECTURE 02:** *Mutually/ Non Mutually Exclusive Events*


### Mutually Exclusive (Disjoint)

Two events are **mutually exclusive** if they cannot occur at the same time.

1. A person cannot run forward and backward at the same time.
2. A person cannot stand and sit at the same time.
3. coin cannot have head and tail at the same time

#### P(A U B) = P(A) + P(B)


exclusive.PNG

```
P(head or tail) = P(head) + P(tail) = 0.5 + 0.5 =1
```

### Non-Mutually Exclusive (Disjoint)

Two events are **non-mutually exclusive** if they CAN occur at the same time.

> 1. A person can run and sweat at the same time.
> 2. A person can stand and wave at the same time.
> 3. A number on a die is odd or divisible by 3 at the same time.

#### P(A U B) = P(A) + P(B) - P(A ∩ B)

```
P(odd or divisible by 3) = P(odd) + P(divisible by 3) - P(odd and divisible by 3) = (3/6) + (2/6) - (1/6) = 4/6

## odd: 
## divisible by 3: 

1,
```

## **LECTURE 03:** *Independent/Dependent Events*

Events are **independent** if the occurrence of one event doesn’t influence the other.
> 1. Raining and scoring A+
> 1. Tossing a head and drawing a heart

### P(A ∩ B) = P(A).P(B)

Events are **dependent** if the occurrence of one event does influence the other.
> 1. Raining and flooding
> 2. Drawing a heart and drawing a heart again

### P(A ∩ B) = P(A).P(B|A)  

We express the probability of event A occurring, given event B has occurred the following way P(B|A).We call this the conditional probability.

P(B|A) is also called **conditional probability**

Just as there are three rules of probability, there are three types of probability, too.

Marginal probability: The likelihood of a single event occurring, independent of any other events happening before or after it.

Joint probability: The likelihood of two random events happening together, where the occurrence of one does not affect the occurrence of the other.

Conditional probability: The likelihood of two events happening together, where the occurrence of one affects the occurrence of the other.

### **CONDITION PROBABILITY**

P(B|A) is Probability of event B given A has already occured

### P(B|A) = P(A ∩ B)/P(A)

conditional probability.PNG

**Question:**
In a group of 100 sports car buyers, 40 bought alarm systems, 30 purchased bucket seats, and 20 purchased an alarm system and bucket seats. If a car buyer chosen at random bought an alarm system, what is the probability they also bought bucket seats?

### Answer:

P(A)= It’s given in the question as 40%, or 0.4

P(A∩B)= This is the intersection of A and B: both happening together. It’s given in the question 20 out of 100 buyers, or 0.2.

P(B|A) = P(A∩B) / P(A) = 0.2 / 0.4 = 0.5.
