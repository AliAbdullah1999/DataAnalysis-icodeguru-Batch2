```python
mean_m=data.groupby('Global Category')['Morningstar Rating'].mean()
mean_m
```

```python
def remove_nan(row):
   if pd.isnull(row['Morningstar Rating']):
      return mean_m[row['Global Category']]
   else:
        return row['Morningstar Rating']

        # How we return Integer values from user define function     (Type casting of user fn)
```

```python
data['Morningstar Rating_new']=data.apply(remove_nan,axis=1)
```

```python
data[data['Morningstar Rating_new'].isnull()]
```

```python
mean=data['Morningstar Rating_new'].mean()
mean
```

```python
def removenan_again(row):
   if pd.isnull(row['Morningstar Rating_new']):
     return mean
   else:
      return row['Morningstar Rating_new']
```

```python
data['Morningstar Rating_new']=data.apply(removenan_again,axis=1)
```
